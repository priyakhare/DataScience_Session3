```python
#                                            Session3 Assignment 



#***********************************************************************************************************

#   Program 1.1 : To Implement myreduce() function thats works like Pythons built-in function reduce()

#***********************************************************************************************************
 
# Function MyReduce takes two inputs, function name and Sequence(list)) as input. Like in reduce Functionality 
# myreduce takes two list items in order at a time and invokes the function given.
# In the First call , it takes first 2 items of sequence executes the function, then the returned Value of the 
# Function is considered one of the values of next iteration of For loop and the second value is the next Seqence Item , in this way function is called till the last sequence item reached and 
# returned value this time is the output of the reduce functionality , that is returned by myreduce function

def myreduce(funcName , funclst):
    a = funclst[0]
    for i in range(0,len(funclst)-1):
        b = funclst[i+1]
        a = funcName(a,b)
    return a
    

# Examples of Functions to use ,and show how the myreduce function works exactly  like reduce(in built Python) function
max_find = lambda a,b: a if (a > b) else b
min_find  = lambda a,b: a if (a < b ) else b

# Setting up the Function name (in this case max_find - to find maximum number) and Input Sequence to be worked on 
funclst = [12,23,14,45,23,78,33,100]
funcName = max_find

# Invoking the myreduce function  and printing the reduced Value based on logic of function 
word = myreduce(funcName,funclst) 
print("The reduced Value  is {}".format(word))    
    
#***********************************************************************************************************************    
```

```python
#************************************************************************************************************    
# Program 1.2  :  Write a Python program to implement your own myfilter() function which works exactly 
#                 like Python's built-in function filter() 
#*************************************************************************************************************


# Implementing Filter(Python inbuilt) in myFilter function , that takes twon inputs (function to be called and Sequence
# For Each Iterable value , calls the function and executes the functionality and filters only those values which return
# true .

def myfilter(funcName, lst):
    l = []
    for i in lst:
        if funcName(i) == True:
            l.append(lst[i])
    return l

# Function to filter out the odd numbers from the list 
def odd_check(num):
    if num%2 == 0:
        return False
    else:
        return True
    
    
# Setting the input Values for myFilter Function , i.e function name and Sequence to be filtered    
lst =range(20)
funcName = odd_check

#Invoking the myfilter function with the function and Sequence and printing the list

lstfiltered = list(myfilter(funcName , lst))
print(lstfiltered)
#*************************************************************************************************************
```

```python
# *******************************************************************************************************
# Program 3 :   Implement a function longestWord() that takes a list of words and returns the longest one. 
    
# ********************************************************************************************************    


# function longestWord , Compares 2  members at a time based on max lenght and thereafter compares the max obtained with the next
# member in lits  

# Implementing function in traditional way  
def LongestWord(lst):
    maxln = 0
    for word in lst:
        wordln = len(word)
        if wordln >= maxln:
            maxln = wordln
            lnword = word
    return lnword      


wordLst = ['aasda','ad','afg','fsfsdfsfsfsfs','sdasdad','dfsd','ZdsfsdfsdfsdfsdfZdfdfsd']


# Implementing the longest word Function by using lambda

# from functools import reduce
# lngword =str(reduce(lambda a,b : a if (len(a) > len(b)) else b ,wordLst))

lngword = LongestWord(wordLst)

print("The Longest Word in the list is '" + lngword +"'")

#**********************************************************************************************************
```
